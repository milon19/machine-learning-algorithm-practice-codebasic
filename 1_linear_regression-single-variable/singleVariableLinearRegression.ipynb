```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('arer(sq-ft)')
plt.ylabel('price(USD$')
plt.title('Home Prices according to Area')
plt.scatter(df.area, df.price, color="red", marker="+")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price) #area = x, price = y fit(x, y)
```

```python
reg.predict([[3300]])
```

**Calculate the data using y = m*x + c**

```python
m = reg.coef_ #m
c = reg.intercept_ #c
print("m = {}, c = {}".format(m, c))
x = 3300
y = m*x + c
y
```

```python
%matplotlib inline
plt.xlabel('arer(sq-ft)')
plt.ylabel('price(USD$')
plt.title('Home Prices according to Area')
plt.scatter(df.area, df.price, color="red", marker="+")
plt.plot(df.area, reg.predict(df[['area']]))
```

**Predict prices from csv file**

```python
d = pd.read_csv('areas.csv')
d.head(5)
```

```python
prices = reg.predict(d)
d['prices'] = prices
d.head(5)
```

**Save Result to the csv**

```python
d.to_csv("result.csv", index=False)
```
